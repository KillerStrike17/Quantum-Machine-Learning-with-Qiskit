```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

Defining the qbit input and output variables

```python
n = 8
n_qb = 8 # input bits
n_ob = 8# Output Bits
```

Way to create a Circuit

```python
qc_circuit = QuantumCircuit(n_qb,n_ob)
```

Add a check on all the input variables

```python
for j in range(n):
    qc_circuit.measure(j,j)
```

Visualizing the Circuit

```python
qc_circuit.draw(output='mpl')
```

```python
counts = execute(qc_circuit,Aer.get_backend("qasm_simulator")).result().get_counts()
```

```python
plot_histogram(counts)
```
